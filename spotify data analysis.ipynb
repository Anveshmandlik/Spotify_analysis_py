```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds_tracks=pd.read_csv(r'C:\Users\ANVESH\Downloads\archive (2)\tracks.csv')
ds_tracks.head()
```

```python
most_popular = ds_tracks.query('popularity>90', inplace=False).sort_values('popularity', ascending=False)
most_popular[:10]
```

```python
#converting release date to index as date time 
ds_tracks.set_index("release_date", inplace=True)
ds_tracks.index = pd.to_datetime(ds_tracks.index)
ds_tracks.head()
```

```python
ds_tracks[["artists"]].iloc[18]
```

```python
ds_tracks["duration"]=ds_tracks["duration_ms"].apply(lambda x: round(x/1000))
ds_tracks.drop("duration_ms",inplace=True,axis=1)
```

```python
ds_tracks.duration.head()
```

```python
#creating a correlation map matrix
corr_ds=ds_tracks.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_ds,annot=True,fmt=".1g", vmin=-1, vmax=1, center=0, cmap="viridis", linewidth=1, linecolor="Black")
heatmap.set_title("Correlation Heatmap Beteen variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

#energy and acousticness have the most negative corr.and most positive with energy and loudness

```python
#creating new coloumn year from release date coloumn
ds_tracks['dates']=ds_tracks.index.get_level_values('release_date')
ds_tracks.dates=pd.to_datetime(ds_tracks.dates)
years=ds_tracks.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist").set(title="No. of songs per year")
```

```python
#Duration of songs over the years

total_dr= ds_tracks.duration
fig_dim = (18,7)
fig, ax = plt.subplots(figsize = fig_dim)
fig = sns.barplot(x=years, y=total_dr, ax=ax, errwidth=False).set(title="Year vs Duration")
plt.xticks(rotation=90)
```

```python
#Line plot to analyse duration of songs over the years
total_dr = ds_tracks.duration
sns.set_style(style="whitegrid")
fig_dim=(10,5)
fig, ax = plt.subplots(figsize=fig_dim)
fig=sns.lineplot(x=years, y=total_dr, ax=ax).set(title="year vs duration")
plt.xticks(rotation=60)
```
